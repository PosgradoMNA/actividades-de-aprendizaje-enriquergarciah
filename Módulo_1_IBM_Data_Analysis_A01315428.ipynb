```python
#Importación de datos a Python
import pandas as pd
import numpy as np
ruta="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df=pd.read_csv(ruta,header=None)
df.head(5) #Para visualizar las primeros 5 registros del dataset
```

```python
#Actualizar los nombres de las columnas
headers=['symboling','normalized-losses','make','fuel-type','aspiration','num_of_doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height',
         'curb-weight','engine-type','num_of-cylinder','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
df.columns=headers
df.head()
```

```python
df.tail(5) #revisar las últimas 5 observaciones
```

```python
#Remplazamos el '?' con información vacía (NaN)
df1=df.replace('?',np.NaN)
df1.head(5)
```

```python
#Eliminamos los missing values de la columna Price
df1=df1.dropna(subset=["price"], axis=0)
df1.head(20)
df1.shape #Se eliminaron 4 valores
```

```python
df1.columns
```

```python
#Cálculo de estadísticos básicos
#Entendimiento de los datos antes de empezar cualquier análisis
#Esto usualmente se hace para, identificar valores faltantes por cada variable, valores atípicos (outliers) y compatibilidad de la información (no mezclar letras y números y después querer realizar alguna operación matemática)
df1.dtypes #Con esto nos da un catálogo de que tipo de dato está en cada una de las variables de dataset, con esto podemos ver que auqneu algunos variables deben de ser numéricos, actualmente la información está capturada como caracter
```

```python
df1.describe() #Devuelve los estaísticos básicos (contador, media, desviación estándar, valores mínimos y máximos, así como los cuartiles y el rango de clase) de las variables numéricas
```
